```python
a = list()

for i in range(1,8):
    a.append(i/0.081)

plt.figure(figsize=(10,10))

for i in a:
    x = str(i)
    strin = list(x)
    strin.remove('.')
    plt.plot(strin)
```

Dividing any number by 8.1 * x or x^-1 gives numbers that have repeating decimals.

Notably:
 
- 1, 10, 100, ... give sequences that go like: 123456790 (excluding 8) and then repeating
- 2, 11, ... give sequences that interchange between even and odd numbers. It starts like: 246, then 9 (adds 3), then 0, then goes to 135, then 8(adds 3), then 0, and then repeating
- 3, 6, 9, 12, give repeating sequences
- Interestingly, all numbers have a 0 at the 9th place (or 8th if you count from 0)
